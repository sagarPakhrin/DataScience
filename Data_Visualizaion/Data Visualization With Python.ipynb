```python
data = pd.read_csv("countries.csv")
data
```

```python
#compare pupulation growth in the US and chine
```

```python
us = data[data['country'] == 'United States']
```

```python
china = data[data['country'] == 'China']
china
```

```python
plt.plot(us.year,us.population/10**6, label="Us")
plt.plot(china.year,china.population/10**6, label="China")
plt.xlabel("Year")
plt.ylabel("Population in Million")
plt.title("Percentage growth of Us and China")
plt.legend()
```
