```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
```

```python
#creating multiplte lines in the same figure
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x,np.sin(x-0), color='blue')        #specify color by name
plt.plot(x,np.sin(x-1), color='g')           #short color code (rbgcmyk)
plt.plot(x,np.sin(x-2), color='0.57')        #Grayscale between 0 and 1
plt.plot(x,np.sin(x-3), color='#ffdd44')     #Hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x-4), color=(1.0,0.2,0.3)) #RGB tuple, values 0 and 1
plt.plot(x,np.sin(x-5), color='chartreuse')  #all the HTML color names support
```

###### Adjusting line style using linestyle keywords

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

#for short you can use the following codes:
plt.plot(x, x+4, linestyle='-')   #solid
plt.plot(x, x+5, linestyle='--')  #dashed
plt.plot(x, x+6, linestyle='-.')  #dashdot
plt.plot(x, x+7, linestyle=':')   #dotted
```

###### Combining linestyle and color code

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1,'--c')
plt.plot(x, x+2,'-.k')
plt.plot(x, x+3,':r')
```

##  Adjusting the Plot: Axes limit

```python
plt.plot(x,np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
#reversing the graph
plt.plot(x,np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

## Labeling Plots

```python
plt.plot(x,np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sine(x)")
```

```python
plt.plot(x,np.sin(x), '-g', label="sin(x)")
plt.plot(x,np.cos(x), ':b', label="cos(x)")
plt.axis('equal')
plt.legend()
```

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10), ylim(-2,2),
      xlabel='x',ylabel='sn(y)')
```
